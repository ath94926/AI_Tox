```python
import pandas as pd
import numpy as np
import pubchempy as pcp
from tqdm import tqdm
```

```python
data = pd.read_csv('EXPORT_LVL5&6_ASID7_TOX21_200730.csv', low_memory=False)
```

```python
data.pivot_table('hitc','casn','aenm')
```

```python
df = data.pivot_table('hitc','casn','aenm')
df.index
```

```python
smiles_list = []

for i, casn in tqdm(enumerate(df.index)):
    try:
        cmpnd = pcp.get_compounds(casn, 'name')
        smiles_list.append(cmpnd[0].isomeric_smiles)
    except:
        smiles_list.append(float('nan'))
```
